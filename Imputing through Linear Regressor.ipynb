```python
data = pd.read_csv('Excel.xlsx')
data
```

```python
columns_to_impute = ['subscribers_for_last_30_days', 'video_views_for_the_last_30_days']

numeric_features = ['rank', 'subscribers','video views', 'uploads', 'country_rank', 'channel_type_rank',
                    'lowest_monthly_earnings', 'highest_monthly_earnings','Population', 
                    'Gross tertiary education enrollment (%)']
categorical_features = ['category', 'Country', 'channel_type', 'created_year']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),  # StandardScaler for numeric features
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features)  # OneHotEncoder for categorical
    ])
```

```python
data.isnull().sum()
```

```python
data[data[column].isna()]
```

```python
for column in columns_to_impute:
    # Create a subset of data where the target column is not missing
    non_missing_data = data[data[column].notna()]
    
    y = non_missing_data[column]    
    X = non_missing_data[numeric_features + categorical_features]
    
    # Apply preprocessing (scaling + one-hot encoding --> handle_ignore to ingore previous fuckups of imbalanced dataset)
    # to the features
    X_encoded = preprocessor.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=41)
    
    # Initialize and train the linear regression model
    regressor = LinearRegression()
    regressor.fit(X_train, y_train)
    
    # Predict missing values
    missing_data = data[data[column].isna()]
    print('OKAY SO FAR')
    if not missing_data.empty:
        X_missing = missing_data[numeric_features + categorical_features]
        X_missing_encoded = preprocessor.transform(X_missing)
        missing_data[column] = regressor.predict(X_missing_encoded)
        
        # Impute the missing values in the original dataset
        data.loc[data[column].isna(), column] = missing_data[column]
    print('WOHOOOOO WORSHIP ME PEASANTS')
```
